### Total COVID Cases of Multiple Countries

```python
b_GER = df_country_info['location'] == 'Germany'
b_US = df_country_info['location'] == 'United States'
b_INDIA = df_country_info['location'] == 'India'
b_ITALY = df_country_info['location'] == 'Italy'
```

```python
b_all = b_GER|b_US|b_INDIA|b_ITALY
```

```python
df_my_list = df_country_info[b_all]
df_my_list = df_my_list.reset_index(drop=True)
#df_my_list.head(5)
```

```python
fig = px.line(df_my_list, x="date", y="total_cases", color='location', title='COVID Cases')
fig.show()
```

### The Relative Cases Overtime of Covid Infectors (absolute Covid cases/population size)

```python
df_my_list['total_cases_norm'] = df_my_list['total_cases']/df_my_list['population']
#df_my_list[['date', 'total_cases', 'location','total_cases_norm', 'population']]
```

```python
fig = px.line(df_my_list, x="date", y="total_cases_norm", color='location', title='Relative COVID Cases')
fig.show()
```

### Save the Graph as an Image

```python
fig.write_image("../images/Relative_COVID_Cases.png")
```

### The vaccination rate (percentage of the population) over time

```python
df_my_list['vaccination_rate'] = (df_my_list['people_fully_vaccinated']/df_my_list['population']) * 100
df_my_list = df_my_list.loc[(df_my_list["date"] >= "2020-12-13")]
```

```python
#df_my_list[['date', 'total_cases', 'location','vaccination_rate', 'people_fully_vaccinated', 'population']]
```

```python
fig = px.line(df_my_list, x='date', y="vaccination_rate", color='location', title='Vaccination Rate')
fig.show()
```

```python
fig.write_image("../images/Vaccination_Rate.png")
```
